```python
#shaping in np array
import numpy as np
arr=np.array([[3,6,2,5],[7,8,0,4]])
print(arr.shape)

arr1=np.array([
    [[45,64,86],[893,22,37]],
    [[87,99,45],[62,23,76]],
    [[23,66,87],[11,23,44]]
])
print(arr1.shape)

# to reshape

arr2=np.array([3,4,6,7,8,1,8,99,55,66])   # to split
a=arr2.reshape(5,2)
print(a)

arr3=np.array([[54,77,54],[99,37,67]])  # to combine
a1=arr3.reshape(-1)
print(a1)

arr4=np.array([1,2,3,4])
x=np.concatenate((arr4,a1))
print(x)
 
```
